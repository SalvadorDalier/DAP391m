# 📦 Online Retail Dataset – Step 1: Data Import & Overview

**Mục tiêu:** Import dataset, xác định cấu trúc dữ liệu, phát hiện vấn đề sớm để lập kế hoạch cleaning.

**Dataset:** Online Retail (2009–2010)  
**Columns:**
- `InvoiceNo` – Mã hóa đơn
- `StockCode` – Mã sản phẩm
- `Description` – Mô tả sản phẩm
- `Quantity` – Số lượng mua
- `InvoiceDate` – Ngày hóa đơn
- `UnitPrice` – Giá đơn vị
- `CustomerID` – Mã khách hàng
- `Country` – Quốc gia

**Goal:** Phân tích xu hướng bán hàng online retail

---
## 🔧 CELL 1: Setup & Import

**Mục đích:** Import thư viện, upload file từ máy local lên Colab, load vào DataFrame, xác định và convert kiểu dữ liệu cho từng cột.

```python
# ============================================================
# CELL 1: SETUP & IMPORT
# ============================================================

# --- 1.1 Import all necessary libraries ---
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
from google.colab import files

print("✅ Libraries imported successfully!")
print(f"   pandas  : {pd.__version__}")
print(f"   numpy   : {np.__version__}")
print(f"   seaborn : {sns.__version__}")

# --- 1.2 Upload file from local machine ---
print("\n📂 Please upload your Excel/CSV file:")
uploaded = files.upload()

# --- 1.3 Load data into a pandas DataFrame ---
filename = list(uploaded.keys())[0]
print(f"\n📄 File uploaded: {filename}")

if filename.endswith('.csv'):
    df = pd.read_csv(filename, encoding='unicode_escape')
elif filename.endswith(('.xlsx', '.xls')):
    df = pd.read_excel(filename)
else:
    raise ValueError("❌ Unsupported file format. Please upload a CSV or Excel file.")

print(f"✅ Data loaded successfully! Shape: {df.shape}")

# --- 1.4 Define categorical and numerical columns ---
# Note: Adjust column names below to match your actual file headers exactly.
categorical_cols = ['InvoiceNo', 'StockCode', 'Description', 'InvoiceDate', 'Country']
numerical_cols   = ['Quantity', 'UnitPrice', 'CustomerID']

# Define lists for categorical and numerical columns (assuming example column names)
# **Note:** You will need to adjust these column names based on your specific CSV file.
# categorical_cols = ['categorical_column_1', 'categorical_column_2']
# numerical_cols   = ['numerical_column_1', 'numerical_column_2']

# Categorical columns = ['InvoiceNo', 'StockCode', 'Description', 'InvoiceDate', 'Country']
# Numerical columns   = ['Quantity', 'UnitPrice', 'CustomerID']

# --- 1.5 Convert categorical columns to 'category' dtype ---
for col in categorical_cols:
    if col in df.columns:
        df[col] = df[col].astype('category')
    else:
        print(f"⚠️  Column not found (skipped): '{col}'")

# --- 1.6 Convert numerical columns to numeric dtype (errors='coerce') ---
for col in numerical_cols:
    if col in df.columns:
        df[col] = pd.to_numeric(df[col], errors='coerce')
    else:
        print(f"⚠️  Column not found (skipped): '{col}'")

# --- 1.7 Print identified columns ---
print("\n📊 Column Classification:")
print(f"   Categorical columns : {categorical_cols}")
print(f"   Numerical columns   : {numerical_cols}")

# --- 1.8 Display DataFrame info ---
print("\n📋 DataFrame Info:")
print("-" * 50)
df.info()
```

```python
# ============================================================
# CELL 2: DATA OVERVIEW
# ============================================================

# --- 2.1 Shape of the DataFrame ---
print("=" * 60)
print("📐 SHAPE")
print("=" * 60)
rows, cols = df.shape
print(f"   Rows    : {rows:,}")
print(f"   Columns : {cols}")

# --- 2.2 DataFrame Info ---
print("\n" + "=" * 60)
print("📋 DATAFRAME INFO")
print("=" * 60)
df.info()

# --- 2.3 Descriptive Statistics ---
print("\n" + "=" * 60)
print("📈 DESCRIPTIVE STATISTICS")
print("=" * 60)
display(df.describe(include='all'))

# --- 2.4 Data Types ---
print("\n" + "=" * 60)
print("🔠 DATA TYPES (dtypes)")
print("=" * 60)
print(df.dtypes)

# --- 2.5 First 5 Rows ---
print("\n" + "=" * 60)
print("👀 FIRST 5 ROWS")
print("=" * 60)
display(df.head())

# --- 2.6 Column Names ---
print("\n" + "=" * 60)
print("🏷️  COLUMN NAMES")
print("=" * 60)
print(df.columns.tolist())

# --- 2.7 Missing Values per Column ---
print("\n" + "=" * 60)
print("❓ MISSING VALUES PER COLUMN")
print("=" * 60)
missing = df.isnull().sum()
missing_pct = (df.isnull().sum() / len(df) * 100).round(2)
missing_df = pd.DataFrame({
    'Missing Count' : missing,
    'Missing (%)'   : missing_pct
})
missing_df = missing_df[missing_df['Missing Count'] > 0].sort_values('Missing Count', ascending=False)

if missing_df.empty:
    print("   ✅ No missing values found!")
else:
    display(missing_df)
    print(f"\n   ⚠️  Total missing cells : {df.isnull().sum().sum():,}")
    print(f"   ⚠️  Columns with nulls  : {(df.isnull().sum() > 0).sum()}")
```
